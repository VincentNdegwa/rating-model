```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

model = joblib.load("imdb_rating_prediiction_model.pkl")

new_data = pd.DataFrame({
    'Year': [2023.0, 2023.0, 2023.0, 2023.0, 2023.0],
    'Duration (min)': [116, 145, 175, 97, 98],
    'Metascore': [82, 37, 39, 30, 44],
    'Votes': ['10,870', '20,101', '45,805', '47,605', '45,186'],
    'Review Count': ['247', '364', '178', '456', '418'],
})


new_data['Votes'] = new_data['Votes'].replace({',': ''}, regex=True).astype(float)
new_data['Review Count'] = new_data['Review Count'].replace({',': ''}, regex=True).astype(float)

predictions = model.predict(new_data)
new_data['Predicted Rating'] = predictions


plt.figure(figsize=(10,6))
sns.barplot(x = new_data.index, y= "Predicted Rating",data=new_data)

plt.title("Predicted Ratings for Movies", fontsize=16)
plt.xlabel("Movie Index", fontsize=12)
plt.ylabel("Predicted Rating", fontsize=12)
plt.xticks(rotation=45)

# Display the plot
plt.show()

```

```python
plt.figure(figsize=(10,6))

sns.lineplot(x='Duration (min)', y='Predicted Rating', data=new_data)

plt.title("Duration vs Predicted Rating", fontsize=16)
plt.xlabel("Duration (min)", fontsize=12)
plt.ylabel("Predicted Rating", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Votes', y='Predicted Rating', data=new_data,)

# Adding titles and labels
plt.title("Votes vs Predicted Rating", fontsize=16)
plt.xlabel("Votes", fontsize=12)
plt.ylabel("Predicted Rating", fontsize=12)

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Votes', y='Predicted Rating', data=new_data, hue='Predicted Rating', palette='coolwarm', s=100)

# Adding titles and labels
plt.title("Votes vs Predicted Rating", fontsize=16)
plt.xlabel("Votes", fontsize=12)
plt.ylabel("Predicted Rating", fontsize=12)

# Display the plot
plt.show()
```
